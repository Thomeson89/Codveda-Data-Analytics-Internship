# Data Cleaning

```python
# Check for missing value

df.isnull().sum()
```

Fill missing categorical data with unknown(location, payment method, item)

```python
df['Location'] = df['Location'].fillna('Unknown')
df['Payment Method'] = df['Payment Method'].fillna('Unknown')
df['Item'] = df['Item'].fillna('Unknown')

print(' categorical columns fixed and filled')
```






change  the data type of quantity, price per unit and total spent to numeric before filling

```python
# Change data formart of Quantity, Price and Total Spent to numeric
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors = 'coerce')
```

```python
# fill the missing values in quantity and price column  with the median to preserve data size
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
```

```python
# fill Total Spent from quantity and price using total spent = quantity * price per unit
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])
```

```python
print( 'numeric column fixed and filled')
```

Clean the date column


```python
# Convert to datetime
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
# Drop rows where the date is still missing
df.dropna(subset=['Transaction Date'], inplace=True)

print(' date column fixed and filled')
```

```python
# Rename Column

df.columns = df.columns.str.lower().str.replace(" ", "_")
```
